### Grouping movies based on Rating(Excellent,Very Good, Good, Average, Poor)

```python
# we will be doing this using functions
def rating(rating):
    if rating >= 8.0:
        return 'Excellent'
    elif rating >= 7.0:
        return 'Very Good'
    elif rating >= 6.0:
        return 'Good'
    elif rating >= 5.0:
        return 'Average'
    else:
        return 'Poor'
        
```

```python
data['Class'] = data['Rating'].apply(rating)
```

lets get a count of the differnet classes

```python
Movie_class = data.groupby('Class')['Rating'].count().sort_values()
Movie_class
```

```python
sns.countplot(x = 'Class', data = data)
```
